```python
#Q1, PART B
X = df.drop(['Unnamed: 0','0'], axis = 1)
X = np.matrix(X)
Y = df.as_matrix(["0"])
```

```python
model_lr = linear_model.LinearRegression()
model_lr.fit(X,Y)
coeff= model_lr.coef_
beta0= model_lr.intercept_[0]
```

```python
#Equation to find predicted values of yi= beta0 + xi1*beta1 + xi2*beta2 + ....
Y_predicted = beta0 + np.dot(X, coeff.T)

#Equation to find variance of noise = (sum(Y[i]-Y_predicted[i])**2)/n-2
sum=0.0
for i in range(len(Y)):
    sum=sum+(Y[i]- Y_predicted[i])**2
Variance=sum/(len(X)-2)
print "Estimate of Variance of the noise: ",Variance
```

```python
#Q2, Part A
#Download the data set midterm-data-set-2. Fit a least-squares solution and report the
#error of the fit (i.e., the magnitude of the residual: ||kXB - y||
df2 = pd.read_csv('midterm_data_set_2.txt')
df2.head(n=3)
```

```python
X=df2[df2.columns[-1:]]
Y=df2[df2.columns['0']]
```

```python
plt.plot(X,Y,'g*')
plt.grid(True)
plt.ylabel("Y",fontsize=12)
plt.xlabel("X",fontsize=12)
plt.title("Q2")
plt.xlim(-3, 3)
plt.ylim(-5, 105)
plt.show()
```

```python
#Q2 Part a
# http://docs.scipy.org/doc/numpy-1.10.0/reference/generated/numpy.linalg.lstsq.html
# B1 = m; B0 = c
# y = B1x + B0
# y = Ap, where A = [[x 1]] and p = [[B1], [B0]]
x = df2["1"]
A = np.vstack([x, np.ones(len(x))]).T
#print A
y = np.array(Y)
#m, c = np.linalg.lstsq(A, y)[0]
#B1, B0 = np.linalg.lstsq(A, y)[0]
model, resid = np.linalg.lstsq(A, y)[:2]
B1, B0 = model
#np.linalg.lstsq(A, y)
print("m(B1) is: {}.".format(round(B1,8)))
print("c(B0) is: {}.".format(round(B0,8)))
r2 = 1 - resid / (y.size * y.var())
print("R2 is: {}.".format(round(r2[0],8)))
```

```python
# PLOT data with least squares solution
# Add bestfit line to ScatterPlot using B1 and B0 values
# y = B1x + B0
part_b, = plt.plot(X,Y,'g*')
plt.grid(True)
plt.ylabel("Y",fontsize=12)
plt.xlabel("X",fontsize=12)
plt.title("Q2")
plt.xlim(-3, 3)
plt.ylim(-5, 105)
xVals = np.arange(-3, 4)
yVals = (B1*xVals) + B0
plt.plot(xVals,yVals,'r-',linewidth=2.0)
plt.show()
```

The scatter plot and the Least Squares linear fit visually shows that the linear fit is poor.  Need to look for non-linear transformation.  The R2 value is very close to zero which meant the straight linear model was a poor fit

```python
# Q2 - Part B
# Try a Log-Log Transformation
logX = np.log(df2["1"])
logY = np.log(df2["0"])

A = np.vstack([logX, np.ones(len(logX))]).T
#print A
y = np.array(logY)
#m, c = np.linalg.lstsq(A, y)[0]
loglogB1, loglogB0 = np.linalg.lstsq(A, y)[0]
print("B1 is: {}.".format(round(loglogB1,8)))
print("B0 is: {}.".format(round(loglogB0,8)))
r2 = 1 - resid / (y.size * y.var())
print("R2 is: {}.".format(round(r2[0],8)))
```

```python
# Q2 - Part B
# Plot a Log-Log Transformation

plt.plot(logX,logY,'k*')
plt.grid(True)
plt.ylabel("log( Y )",fontsize=12)
plt.xlabel("log( X )",fontsize=12)
plt.title("log(Y) vs log(X)")
plt.xlim(-10, 10)
xVals = np.arange(-9, 9)
yVals = (loglogB1*xVals) + loglogB0
plt.plot(xVals,yVals,'r-',linewidth=2.0)
plt.show()
```

```python
# Q2 - Part B
# Try a Linear-Log Transformation
linX = df2["1"]
logY = np.log(df2["0"])

A = np.vstack([linX, np.ones(len(logX))]).T
#print A
y = np.array(logY)
#m, c = np.linalg.lstsq(A, y)[0]
loglinB1, loglinB0 = np.linalg.lstsq(A, y)[0]
print("B1 is: {}.".format(round(loglinB1,8)))
print("B0 is: {}.".format(round(loglinB0,8)))
r2 = 1 - resid / (y.size * y.var())
print("R2 is: {}.".format(round(r2[0],8)))
```

```python
# Q2 - Part B
# Plot a Linear-Log Transformation

plt.plot(linX,logY,'go')
plt.grid(True)
plt.ylabel("log Y",fontsize=12)
plt.xlabel("X",fontsize=12)
plt.title("X vs log Y")
plt.xlim(-3, 3)
xVals = np.arange(-10, 10)
yVals = (loglinB1*xVals) + loglinB0
plt.plot(xVals,yVals,'r-',linewidth=2.0)
plt.show()
#Good plot with a linear-Log relationship  
```
